```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/florida_bridge_cleaned.csv")
```

```python
condition_cols = [
    "DECK_COND_058",
    "SUPERSTRUCTURE_COND_059",
    "SUBSTRUCTURE_COND_060",
    "CHANNEL_COND_061",
    "CULVERT_COND_062",
    "BRIDGE_CONDITION",
    "LOWEST_RATING"
]
```

```python
for col in condition_cols:
    print(f"\n{'='*40}")
    print(col)
    print(df[col].value_counts(dropna=False).sort_index())
```

```python
for col in condition_cols:
    df[col] = df[col].replace("N", np.nan)
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["MIN_CONDITION"] = df[condition_cols].min(axis=1)
```

```python
def risk_category(score):
    if score <= 4:
        return "High"
    elif score <= 6:
        return "Medium"
    else:
        return "Low"

df["Infrastructure_Risk"] = df["MIN_CONDITION"].apply(risk_category)
```

```python
df["Infrastructure_Risk"].value_counts()
```

```python
df.to_csv("../data/florida_bridge_with_risk_final.csv", index=False)
```
